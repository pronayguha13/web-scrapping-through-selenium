```python
import pandas as pd
```

```python
df = pd.read_csv("player_passes.csv")
```

```python
df
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df
```

# Total number of passes by each team

```python
def get_total_passes():
    team_groups = df.groupby("team")
    total_passes = team_groups["total_passes"].sum()
    s = pd.Series(data=total_passes)
    return s

get_total_passes()
```

# sort the teams by total number of passes

```python
s = get_total_passes()

s.sort_values(ascending=False)
```

# Get player who had the most passes for each team

```python
def get_passmaker():
    pass_maker = df.loc[df.groupby("team")["total_passes"].idxmax()]
    return pass_maker.sort_values(by=["total_passes"], ascending= False)
get_passmaker()
```

# get player who completed most passes for each of team

```python
def get_complete_pass_maker():
    complete_pass_maker = df.loc[df.groupby("team")["completed_passes"].idxmax()]
    
    return complete_pass_maker.sort_values(by=["completed_passes"], ascending=False)

get_complete_pass_maker()
```

# get the completion percentage column

```python
def completion_percentage(row):
    completion_percentage = (
        round((row["completed_passes"] / row["total_passes"]), 4)
        if row["total_passes"] != 0
        else 0
    )

    row["percentage_completed"] = completion_percentage
    return row


df = df.apply(completion_percentage, axis=1)
```

```python
df.sort_values(by=["completed_passes", "total_passes", "percentage_completed"],ascending= False)
```
